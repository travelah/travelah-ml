###**Text Pre-Processing Travel**

```python
import re
from nltk.corpus import stopwords as nltk_stopwords

clean_spcl = re.compile('[/(){}\[\]\|@,;]')
clean_symbol = re.compile('[^0-9a-z #+_]')
stop_words = set(nltk_stopwords.words('english'))

def clean_df(text):
    text = text.lower()
    text = re.sub("[^a-zA-Z0-9 ]", "", str(text))
    text = clean_spcl.sub(' ', text)
    text = clean_symbol.sub('', text)
    text = ' '.join(word for word in text.split() if word not in stopwords)
    return text
```

```python
def search_improv(query):
    query = clean_df(query)
    query_vector = vectorizer.transform([query])
    similarity_scores = cosine_similarity(query_vector, tfidf_hotel).flatten()
    top_indices = similarity_scores.argsort()[::-1][:10]  # Get the indices of the top 10 most similar places
    results = hotel_df.iloc[top_indices]
    return results
```

```python
def search_improv_region(region, query):
    region = clean_df(region)
    query = clean_df(query)
    query_vector = vectorizer.transform([query])

    similarity_scores = cosine_similarity(query_vector, tfidf_hotel).flatten()

    # Split region into individual words
    region_words = region.split()

    # Filter based on region
    region_indices = hotel_df['region'].apply(lambda r: re.search(r'\b{}\b'.format(region), r, flags=re.IGNORECASE) is not None)
    filtered_scores = similarity_scores * region_indices

    top_indices = filtered_scores.argsort()[::-1][:10]
    results = hotel_df.iloc[top_indices]
    return results
```

```python
def search_improv_rating(region, query, rating=None):
    region = clean_df(region)
    query = clean_df(query)
    query_vector = vectorizer.transform([query])

    similarity_scores = cosine_similarity(query_vector, tfidf).flatten()

    # Filter based on region
    region_indices = travel_df['region '].str.contains(region, case=False, regex=False)
    filtered_scores = similarity_scores * region_indices

    if rating:
        if rating == 'low_rating':
            filtered_indices = travel_df[travel_df['rating_google'] < 3.5].index
        elif rating == 'average_rating':
            filtered_indices = travel_df[(travel_df['rating_google'] >= 3.5) & (travel_df['rating_google'] < 4.2)].index
        elif rating == 'high_rating':
            filtered_indices = travel_df[travel_df['rating_google'] >= 4.2].index
        else:
            filtered_indices = travel_df.index
    else:
        filtered_indices = travel_df.index

    filtered_scores[~travel_df.index.isin(filtered_indices)] = 0  # Set filtered scores of non-matching ratings to 0

    top_indices = filtered_scores.argsort()[::-1][:10]  # Get the indices of the top 10 most similar places
    results = travel_df.iloc[top_indices]
    return results
```

##**Model Deep Learning**

```python
import re
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from nltk.corpus import stopwords as nltk_stopwords
from sklearn.utils import shuffle

clean_spcl = re.compile('[/(){}\[\]\|@,;]')
clean_symbol = re.compile('[^0-9a-z #+_]')
stopwords_set = set(nltk_stopwords.words('english'))

def clean_df(text):
    text = text.lower()
    text = re.sub("[^a-zA-Z0-9 ]", "", str(text))
    text = clean_spcl.sub(' ', text)
    text = clean_symbol.sub('', text)
    text = ' '.join(word for word in text.split() if word not in stopwords_set)
    return text

# Shuffle the data
travel_df = shuffle(travel_df)

# Prepare data for travel spots
combined_text = travel_df['combined_text'].apply(clean_df)
spot_labels = travel_df['region'].astype('category').cat.codes

# Prepare data for hotels
hotel_desc = hotel_df['description'].apply(clean_df)
hotel_labels = hotel_df['region'].astype('category').cat.codes

# Tokenize text for travel spots
spot_tokenizer = Tokenizer()
spot_tokenizer.fit_on_texts(combined_text)
spot_vocab_size = len(spot_tokenizer.word_index) + 1
spot_sequences = spot_tokenizer.texts_to_sequences(combined_text)

# Tokenize text for hotels
hotel_tokenizer = Tokenizer()
hotel_tokenizer.fit_on_texts(hotel_desc)
hotel_vocab_size = len(hotel_tokenizer.word_index) + 1
hotel_sequences = hotel_tokenizer.texts_to_sequences(hotel_desc)

# Pad sequences for travel spots
spot_max_sequence_length = max(len(seq) for seq in spot_sequences)
spot_padded_sequences = pad_sequences(spot_sequences, maxlen=spot_max_sequence_length)

# Pad sequences for hotels
hotel_max_sequence_length = max(len(seq) for seq in hotel_sequences)
hotel_padded_sequences = pad_sequences(hotel_sequences, maxlen=hotel_max_sequence_length)

# Convert labels to categorical for travel spots
spot_num_classes = len(np.unique(spot_labels))
spot_labels_categorical = to_categorical(spot_labels, num_classes=spot_num_classes)

# Convert labels to categorical for hotels
hotel_num_classes = len(np.unique(hotel_labels))
hotel_labels_categorical = to_categorical(hotel_labels, num_classes=hotel_num_classes)

# Split data into train and test sets for travel spots
spot_X_train, spot_X_test, spot_y_train, spot_y_test = train_test_split(
    spot_padded_sequences, spot_labels_categorical, test_size=0.2, random_state=42
)

# Split data into train and test sets for hotels
hotel_X_train, hotel_X_test, hotel_y_train, hotel_y_test = train_test_split(
    hotel_padded_sequences, hotel_labels_categorical, test_size=0.2, random_state=42
)

# Define the model for travel spots
spot_embedding_dim = 100
spot_model = Sequential()
spot_model.add(Embedding(spot_vocab_size, spot_embedding_dim, input_length=spot_max_sequence_length))
spot_model.add(LSTM(100, dropout=0.2))
spot_model.add(Dense(spot_num_classes, activation='softmax'))

# Define the model for hotels
hotel_embedding_dim = 100
hotel_model = Sequential()
hotel_model.add(Embedding(hotel_vocab_size, hotel_embedding_dim, input_length=hotel_max_sequence_length))
hotel_model.add(LSTM(100, dropout=0.2))
hotel_model.add(Dense(hotel_num_classes, activation='softmax'))
```

```python
# Compile and train the model for travel spots
spot_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
spot_model.fit(spot_X_train, spot_y_train, validation_data=(spot_X_test, spot_y_test), epochs=100)
```

```python
# Compile and train the model for hotels
hotel_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
hotel_model.fit(hotel_X_train, hotel_y_train, validation_data=(hotel_X_test, hotel_y_test), epochs=20)
```
